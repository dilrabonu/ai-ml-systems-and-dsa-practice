```python
# Scalers
import numpy as np
from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler

X = np.array([[500., 1.], [2000., 3.], [5000., 6.]])

print(StandardScaler().fit_transform(X)) # mean 0, std 1
print(MinMaxScaler().fit_transform(X)) # range [0, 1]
print(RobustScaler().fit_transform(X)) # uses median + IQR, robust to outliers
```
