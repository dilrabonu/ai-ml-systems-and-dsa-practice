```python
import pandas as pd
df = pd.DataFrame({
    'date': pd.date_range('2024-01-01', periods=10),
    'sales': [100, 120, 90, 150, 200, 180, 110, 95, 220, 175]
})

df['rolling_mean_3'] = df['sales'].rolling(window=3).mean()
df['rolling_std_3'] = df['sales'].rolling(window=3).std()
df['rolling_max_3'] = df['sales'].rolling(window=3).max()
df
```

```python
import pandas as pd

df = pd.DataFrame({
    'user_id': ['U1', 'U1', 'U2', 'U2', 'U3', 'U3', 'U1'],
    'timestamp': pd.date_range('2024-01-01', periods=7),
    'amount': [100, 200, 50, 75, 300, 250, 150],
    'merchant': ['Amazon', 'Walmart', 'Amazon', 'Apple', 'Walmart', 'Amazon', 'Apple'],
    'is_fraud': [0, 0, 0, 1, 0, 0, 1]
})

df = df.sort_values(['user_id', 'timestamp'])

df['user_rolling_avg'] = (
    df.groupby('user_id')['amount']
      .rolling(window=3, min_periods=1)
      .mean()
      .reset_index(level=0, drop=True)
)

df
```

```python
df['amount_rolling_sum_7'] = df['amount'].rolling(window=7, min_periods=1).sum()
df['amount_rolling_q95_30'] = df['amount'].rolling(window=30, min_periods=1).quantile(0.95)

df_time = df.sort_values('timestamp').set_index('timestamp')
df_time['amount_rolling_mean_7d'] = df_time['amount'].rolling('7D').mean()

df_time
```
