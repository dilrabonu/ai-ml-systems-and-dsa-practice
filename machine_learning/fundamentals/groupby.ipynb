```python
# groupby
import pandas as pd

df = pd.DataFrame({
    'user_id': ['U1', 'U1', 'U2', 'U2', 'U3', 'U3', 'U1'],
    'amount': [100, 200, 50, 75, 300, 250, 150],
    'merchant': ['Amazon', 'Walmart', 'Amazon', 'Apple', 'Walmart', 'Amazon', 'Apple'],
    'is_fraud': [0, 0, 0, 1, 0, 0, 1]
})
df.head()
```

```python
# basic groupby : sum per user
df.groupby('user_id')['amount'].sum()
```

```python
df.groupby('user_id')['merchant'].value_counts()
```

```python
df.stats = df.groupby('user_id').agg(
    total_spent= ('amount', 'sum'),
    avg_spent=('amount', 'mean'),
    txn_count=('amount', 'count'),
    max_txn=('amount', 'max'),
    fraud_count=('is_fraud', 'sum')
)
df.stats
```

```python
df_columns = df.groupby(['user_id', 'merchant'])['amount'].sum()
df_columns
```

```python
# transform
df_transform = df.groupby('user_id')['amount'].transform('mean')
df_transform
```

```python
df_fraud = df.groupby('user_id')['is_fraud'].sum()
df_fraud
```
