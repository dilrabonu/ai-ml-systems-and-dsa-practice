```python
import numpy as np
import pandas as pd

amounts = pd.Series([0, 10, 100, 1000, 10000, 100000])
log_amounts = np.log1p(amounts)
```

```python
log_amounts
```

```python
# reverse it
np.expm1(log_amounts)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(log_amounts)
plt.show()

# after log
sns.histplot(np.log1p(log_amounts))
plt.show()
```
