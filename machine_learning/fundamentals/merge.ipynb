```python
import pandas as pd

customers = pd.DataFrame({
    'user_id': ['U1', 'U2', 'U3'],
    'name': ['Alice', 'Bob', 'Carol'],
    'age': [30, 45, 28]
})

transactions = pd.DataFrame({
    'user_id': ['U1', 'U1', 'U2', 'U4'],
    'amount': [100, 200, 50, 300]
})
```

```python
transactions
```

```python
df_merged = pd.merge(customers, transactions, on='user_id', how='inner')
df_merged
```

```python
# left join
df_merged_left = pd.merge(customers, transactions, on = 'user_id', how='left')
df_merged_left
```

```python
df_merged_outer= pd.merge(customers, transactions, on='user_id', how='outer')
df_merged_outer
```

```python
df_merged_diff = pd.merge(customers, transactions, left_on='user_id', right_on='user_id')
df_merged_diff
```
